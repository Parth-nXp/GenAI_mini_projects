```python
import torch
import math
import torch.nn.functional as F
```

```python
def scaled_dot_product(q, k, v, d_k):
  scores = torch.einsum('bsd, bcd -> bsc', q, k)
  scaled_scores = scores / math.sqrt(d_k)
  attention_scores = F.softmax(scaled_scores, dim =-1)
  output = torch.einsum('bsc, bsd -> bsd', attention_scores, v)
  return attention_scores, output
```

```python
batch_size = 1
seq_len = 8
d_k = 64

query = torch.randn(batch_size, seq_len, d_k)
key = torch.randn(batch_size, seq_len, d_k)
value = torch.randn(batch_size, seq_len, d_k)

attention_score, output = scaled_dot_product(query, key, value, d_k)
print(attention_score, output)
```
